```python
# importing the machine learning libraries and modules

# pandas, an open-source machine learning library with flexible high-level data structures & analysis tools
import pandas as pd

# numpy(Numerical Python), a popular library supporting large matrices & multi-dimensional data
import numpy as np

# re(Regular Expression), a built-in Python module/package to work with Regular Expressions
import re

# seaborn(built upon Matplotlib), a library to plot graphs & visualize random distributions
import seaborn as sns

# matplotlib, another library used to plot numerical data & plot high-defined figures like pie-charts, histograms, scatterplots, graphs, etc.
# pyplot, an API for Python's matplotlib which help make matplotlib a viable open-source alternative to MATLAB
import matplotlib.pyplot as plt

# style(matplotlib), a matplotlib's built-in package used to style the plots
from matplotlib import style

# using the style for the plot by adjusting it to emulate ggplot 
style.use('ggplot')

# TextBlob, a library used for building programs for text analysis/processing
!pip install textblob      
import textblob      
from textblob import TextBlob

# natural language toolkit (nltk), a Python package used for nlp (natural language processing)
# nltk.tokenize, a module of nltk to tokenize sentences and words
# word_tokenize, a method to split a sentence into words
import nltk
from nltk.tokenize import word_tokenize

# nltk.stem, a package that performs stemming using different classes
# PorterStemmer, one of the classes/algorithms that follows a set of 05 rules for different cases, applied in phases to generate stems
from nltk.stem import PorterStemmer

# nltk.corpus, a module that defines a collection of corpus reader classes, that can be used to access the contents of a diverse set of corpora
# stopwords, a method to remove very common words but don't provide useful information for most text analysis procedures
from nltk.corpus import stopwords

# object that stores the English stopwords (like, 'stop', 'the', 'to', 'and', 'a')
stop_words = set(stopwords.words('english'))

# WordCloud, a Python library to create tag clouds
!pip install wordcloud
import wordcloud
from wordcloud import WordCloud

# CountVectorizer, a method to transform a given text into a vector on the basis of frequency (count) of each word that occurs in the entire text
from sklearn.feature_extraction.text import CountVectorizer

# train_test_split, another method in scikit-learn that splits arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split

# LogisitcRegression, a algorithm used to estimate relationship between a dependent variable and one/more independent variables
from sklearn.linear_model import LogisticRegression

# accuracy_score, a method used to measure ratio of sum of true +ve & true -ve out of all the predictions made
# classification_report, a method to measure quality of predictions from a classification algorithm
# confusion_matrix, a method that evaluate the quality of the output of a classifier on dataset
# ConfusionMatrixDisplay, to create a more interpretable visual display out of table
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
text_df = df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'], axis=1)
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head(10)
```

```python
def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
text_df.head()
```

```python
fig = plt.figure(figsize=(5,5))
sns.countplot(x='sentiment', data = text_df)
```

```python
fig = plt.figure(figsize=(7,7))
colors = ("yellowgreen", "gold", "red")
wp = {'linewidth':2, 'edgecolor':"black"}
tags = text_df['sentiment'].value_counts()
explode = (0.1,0.1,0.1)
tags.plot(kind='pie', autopct='%1.1f%%', shadow=True, colors = colors,
         startangle=90, wedgeprops = wp, explode = explode, label='')
plt.title('Distribution of sentiments')
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

```python
X = text_df['text']
Y = text_df['sentiment']
X = vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("Size of x_train:", (x_train.shape))
print("Size of y_train:", (y_train.shape))
print("Size of x_test:", (x_test.shape))
print("Size of y_test:", (y_test.shape))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(), param_grid)
grid.fit(x_train, y_train)
```

```python
print("Best parameters:", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, y_pred))
print("\n")
print(classification_report(y_test, y_pred))
```
